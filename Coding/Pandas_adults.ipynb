```python
import numpy as np
df.replace("?", np.nan, inplace=True)
df
```

```python
columns_to_drop = ['educational-num', 'capital-gain', 'capital-loss']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
column_distribution = df['age'].value_counts()
print(column_distribution)
```

```python
age=df['age'].between(17, 48, inclusive=True).sum()
print(age)
```

```python
column_distribution = df['workclass'].value_counts()
print(column_distribution)
```

```python
column_distribution = df['education'].value_counts()
print(column_distribution)
```

```python
num1 = df['education'].str.contains('Bachelor', case=False).sum()
num2 = df['education'].str.contains('Masters', case=False).sum()
print(num1+num2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 6))
df.groupby('age')['hours-per-week'].mean().plot(kind='bar')
plt.xlabel('age')
plt.ylabel('hrs/week')
plt.title('Scatter Plot between age and hours-per-week')
plt.show()
```

```python
df['income'].replace({'<=50K': 0, '>50K': 1}, inplace=True)
df
```

```python
result = df.groupby('workclass')['income'].sum()
print(result)
```
